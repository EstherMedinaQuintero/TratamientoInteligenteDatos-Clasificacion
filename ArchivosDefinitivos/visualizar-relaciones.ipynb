```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r"homeLoanAproval_modificado.csv",sep=',')
```

```python
df.info()
```

```python
df.describe()
```

```python
df[df['LoanStatus']=='Y']['ApplicantIncome'].plot.hist(bins=30,alpha=0.5,color='blue', label='LoanStatus=Y')
df[df['LoanStatus']=='N']['ApplicantIncome'].plot.hist(bins=30,alpha=0.5, color='red', label='LoanStatus=N')
plt.legend(fontsize=15)
plt.xlabel("ApplicantIncome score", fontsize=14)
```

```python
df[df['LoanStatus']=='Y']['LoanAmount'].plot.hist(bins=30,alpha=0.5,color='blue', label='LoanStatus=Y')
df[df['LoanStatus']=='N']['LoanAmount'].plot.hist(bins=30,alpha=0.5, color='red', label='LoanStatus=N')
plt.legend(fontsize=15)
plt.xlabel("LoanAmount score", fontsize=14)
```

```python
df[df['LoanStatus']=='Y']['CoapplicantIncome'].plot.hist(bins=30,alpha=0.5,color='blue', label='LoanStatus=Y')
df[df['LoanStatus']=='N']['CoapplicantIncome'].plot.hist(bins=30,alpha=0.5, color='red', label='LoanStatus=N')
plt.legend(fontsize=15)
plt.xlabel("CoapplicantIncome score", fontsize=14)
```

```python
df[df['LoanStatus']=='Y']['LoanAmountTerm'].plot.hist(bins=30,alpha=0.5,color='blue', label='LoanStatus=Y')
df[df['LoanStatus']=='N']['LoanAmountTerm'].plot.hist(bins=30,alpha=0.5, color='red', label='LoanStatus=N')
plt.legend(fontsize=15)
plt.xlabel("LoanAmountTerm score", fontsize=14)
```

```python
sns.jointplot(x='ApplicantIncome',y='LoanStatus',data=df, color='purple', size=12)
```

```python
sns.jointplot(x='LoanAmount',y='LoanStatus',data=df, color='purple', size=12)
```

```python
sns.jointplot(x='CoapplicantIncome',y='LoanStatus',data=df, color='purple', size=12)
```
