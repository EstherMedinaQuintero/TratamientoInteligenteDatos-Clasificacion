```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r"homeLoanAproval_modificado.csv",sep=',')
```

```python
df.info()
```

```python
df.describe()
```

```python
df[df['LoanStatus']=='Y']['ApplicantIncome'].plot.hist(bins=30,alpha=0.5,color='blue', label='LoanStatus=Y')
df[df['LoanStatus']=='N']['ApplicantIncome'].plot.hist(bins=30,alpha=0.5, color='red', label='LoanStatus=N')
plt.legend(fontsize=15)
plt.title ("Histogram of ApplicantIncome score by approved or disapproved credit policies", fontsize=16)
plt.xlabel("ApplicantIncome score", fontsize=14)
```

```python
sns.jointplot(x='ApplicantIncome',y='LoanAmount',data=df, color='purple', size=12)
```
